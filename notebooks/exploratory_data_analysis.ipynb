**Calculating Distances Between Listings and Tourist Areas**

```python
# Function to calculate the distance using geopy
def haversine_distance(latitude1, longitude1, latitude2, longitude2):
    # Radius of Earth in km
    earth_radius_km = 6371

    # Converting longitude and latitude from degrees to radians
    radian_latitude1, radian_latitude2 = np.radians(latitude1), np.radians(latitude2)
    radian_delta_latitude = np.radians(latitude2 - latitude1)
    radian_delta_longitude = np.radians(longitude2 - longitude1)

    # Haversine formula to calculate the great-circle distance
    a = np.sin(radian_delta_latitude / 2)**2 + np.cos(radian_latitude1) * np.cos(radian_latitude2) * np.sin(radian_delta_longitude / 2)**2

    # The distance in km
    distance_km = 2 * earth_radius_km * np.arcsin(np.sqrt(a))
    return distance_km

for area, coordinates in tourist_areas.items():
    df[f"distance_to_{area}"] = df.apply(lambda row: haversine_distance(row["latitude"], row["longitude"], coordinates[0], coordinates[1]), axis=1)
```

**Finding the nearest tourist area**

```python
df["nearest_tourist_area"] = df[[f"distance_to_{area}" for area in tourist_areas]].idxmin(axis=1).str.replace("distance_to_", "")
```

**Average Price by Borough**

```python
# Compare pricing across boroughs and neighborhoods
avg_price_by_borough = df.groupby("neighbourhood_group")["price"].mean()
avg_price_by_neighborhood = df.groupby("neighbourhood")["price"].mean()

print(avg_price_by_borough)

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_by_borough.index, y=avg_price_by_borough.values, hue=avg_price_by_borough.index, palette="viridis", legend=False)
plt.title("Average Price by Borough")
plt.xlabel("Borough")
plt.ylabel("Average Price")

plt.savefig("average_price_by_borough.png", format="png", dpi=300)

plt.show()
```

**Top-Priced Neighborhoods in Each Borough**
